```python
import numpy as np
import pandas as pd 
from time import time
import ast 
```

## Get all keywords that aren't included in the list online

```python
# Get the keywords

key_file = open('keywords.txt', mode='r')
keyword_store = key_file.readlines()
key_file.close()

for i,line in enumerate(keyword_store):
    temp = line.replace('\n', '')
    keyword_store[i] = temp.strip()
 
all_keywords = pd.DataFrame(keyword_store, columns=['tags'])
```

```python
def get_index(tag):
    
    try:
        tag_index = all_keywords[ all_keywords['tags'] == tag ].index[0]
        return tag_index
    except IndexError:
        return None
```

```python
get_index('algebra')
```

```python
questions_all = pd.read_csv('tagged_data.csv', index_col=0)
tags = questions_all['tags'].values
```

```python
for i,tag in enumerate(tags):
    tags[i] = ast.literal_eval(tag)
```

```python
t0 = time()

missing_keywords = {}

for i, tag_set in enumerate(tags):
    for tag in tag_set:
        index = get_index(tag)
        if index == None:
            if tag not in missing_keywords:
                missing_keywords[tag] = 1
            else:
                missing_keywords[tag] += 1
                
t1 = time()

print 'Took ' + str(t1-t0) + ' seconds to find all missing keywords'
```

```python
new_keywords = []

for i,tag in enumerate(missing_keywords.keys()):
    tag = tag.replace('"', '')
    tag = tag.replace('#', '')
    tag = tag.strip()
    new_keywords.append(tag)
```

```python
new_keywords = [t for t in new_keywords if t != '']
new_keywords
```

## Build the total list of keywords

```python
# Get the keywords

key_file = open('keywords.txt', mode='r')
keyword_store = key_file.readlines()
key_file.close()

for i,line in enumerate(keyword_store):
    temp = line.replace('\n', '')
    keyword_store[i] = temp.strip()
 
print 'Total ' + str(len(keyword_store)) + ' keywords'

keyword_store = keyword_store + new_keywords
sorted(keyword_store)

print 'Total ' + str(len(keyword_store)) + ' keywords'

all_keywords = pd.DataFrame(keyword_store, columns=['tags'])
```
