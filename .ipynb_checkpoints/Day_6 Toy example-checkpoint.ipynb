```python
from __future__ import division
import numpy as np
import pandas as pd 
from time import time
import ast 
from sklearn.cross_validation import train_test_split
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.multiclass import OneVsRestClassifier
```

```python
questions_all = pd.read_csv('tagged_data.csv', index_col=0)
questions = questions_all[ ['question'] ].values

# Remove all annoying nan values
remove = []

for i in range(len(questions)):
    if type(questions[i][0]) != str:
        remove.append(i)
        
questions_all = questions_all.drop(questions_all.index[remove])
questions_all = questions_all[ ['question', 'tags'] ]

train, test = train_test_split(questions_all, test_size = 0.2)
```

```python
len(train)
```

```python
# Build the vocabulary of all keywords here
all_keywords = pd.read_csv('all_keywords.txt', index_col=0)
all_keywords = all_keywords['tags'].values.tolist()
```

```python
n_samples = 10000

X_train = []
Y_train = []

for i in range(n_samples):
    #X_train.append(questions_all.iloc[i]['question'])
    #Y_train.append(ast.literal_eval(questions_all.iloc[i]['tags']))
    X_train.append(train[i][0])
    Y_train.append(ast.literal_eval(train[i][1]))
```

```python
X_test = []
Y_test = []

for i in range(n_samples,n_samples + 500):
    X_test.append(questions_all.iloc[i]['question'])
    Y_test.append(ast.literal_eval(questions_all.iloc[i]['tags']))
```

```python
Y_train_indices = []

for tag_list in Y_train:
    y_tags = []
    for tag in tag_list:
        if tag in all_keywords:
            y_tags.append(all_keywords.index(tag))
        
    Y_train_indices.append(y_tags)
```

```python
Y_train = MultiLabelBinarizer().fit_transform(Y_train_indices)
```

```python
classifier = Pipeline([
    ('vectorizer', CountVectorizer(ngram_range=(1,3))),
    ('tfidf', TfidfTransformer()),
    ('clf', OneVsRestClassifier(LinearSVC()))])
```

```python
classifier.fit(X_train, Y_train)
```

```python
y_predict = classifier.predict(X_test)
```

```python
# Get back the indices of the predictions
predicted_labels = []

for i in range(len(y_predict)):
    y_pred = []
    
    for j in range(len(y_predict[i])):
        if y_predict[i][j] == 1:
            y_pred.append(j)   
            
    predicted_labels.append(y_pred)
    
j_scores = []

for i in range(len(predicted_labels)):

    Y_true_num = []
    
    for x in Y_test[i]:
        temp = x.replace('"','')
        Y_true_num.append(all_keywords.index(temp))
    
    A = set(Y_true_num)
    B = set(predicted_labels[0])

    inter = A.intersection(B)
    union = A.union(B)

    j_score = len(inter) / len(union)

    j_scores.append(j_score)
    
j_scores = pd.DataFrame(j_scores)
j_scores.describe()
```
