```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from time import time
import ast 
from __future__ import division
import seaborn as sns
```

```python
def jaccard(intersection, union):
    if len(union) != 0:
        return len(intersection) / len(union)
    else:
        return None
```

```python
def randomly_sample_jaccard(question_tags, n_samples, n_times=1):
    
    jaccard_coefficients = []
    
    for i in range(n_times):
        # Randomly sample
        random_rows = np.random.choice(question_tags.index.values, n_samples)
        random_questions = question_tags.ix[random_rows]

        # Get values as list
        question_text = random_questions['question'].values
        tags = random_questions['tags'].values
        tags_found = random_questions['tags_found'].values

        # Convert to lists
        for i,tag in enumerate(tags):
            tags[i] = ast.literal_eval(tag)

        for i,tag in enumerate(tags_found):
            tags_found[i] = ast.literal_eval(tag)

        # Calculate jaccard score for these randomly chosen values
        jaccard_similarity = []

        for i in range(n_samples):
            y_label = tags[i]
            y_label = set(y_label)

            y_predic = tags_found[i]
            y_predic = set(tags_found[i])

            union = y_label.union(y_predic)
            intersection = y_label.intersection(y_predic)

            jaccard_similarity.append(jaccard(intersection, union))

        random_questions['jaccard'] = jaccard_similarity

        jaccard_coefficients.append(random_questions['jaccard'].mean())
        
    jaccard_coefficients = np.array(jaccard_coefficients)
        
    return jaccard_coefficients.mean()
```

```python
questions_all = pd.read_csv('tagged_data.csv', index_col=0)
question_tags = questions_all[['question', 'tags', 'tags_found']]
```

```python
x = []
y = []

t0 = time()

iterations = 5

n_samples_min = 1000
n_samples_max = 19000
n_samples_step = 1000

for n in range(n_samples_min, n_samples_max, n_samples_step):
    j = randomly_sample_jaccard(question_tags, n, iterations)
    x.append(n)
    y.append(j*100)
    
t1 = time()   

print 'Time taken: ' + str(t1-t0)
```

```python
plt.title('Jaccard Similarity for Naive approach')
plt.xticks(np.arange(n_samples_min, n_samples_max, n_samples_step))
plt.xlabel('Number of questions (randomly chosen)')
plt.ylabel('Average Jaccard similarity score over ' + str(iterations)+ ' iterations')
plt.plot(x,y)
plt.plot([min(x), max(x)],[np.array(y).mean(), np.array(y).mean()])
plt.show()
```
